```python
circles_attributes, circles_targets = make_circles(noise=0.08, factor=0.5, n_samples = 200)
```

```python
plt.scatter(circles_attributes[:, 0], circles_attributes[:, 1], c = circles_targets)
plt.gca().set_aspect('equal')
plt.show()
```

```python
circles_svm = LinearSVC()
```

```python
circles_svm.fit(circles_attributes, circles_targets)
```

```python
circles_svm.score(circles_attributes, circles_targets)
```

```python
circles_attributes_poly = PolynomialFeatures(degree=2).fit_transform(circles_attributes)
```

```python
circles_svm.fit(circles_attributes_poly, circles_targets)
```

```python
circles_svm.score(circles_attributes_poly, circles_targets)
```

```python
circles_attributes_poly.shape # from 2 to 6 features
```

```python
circles_svm_with_kernel.fit(circles_attributes, circles_targets)
```

```python
circles_svm_with_kernel.score(circles_attributes, circles_targets)
```

```python
pusars = pd.read_csv("data/pulsar_star.csv")
```

```python
pusars
```

```python
pusars.target_class.value_counts(normalize=True)
```

```python
pulsars_attribues_train, pulsars_attribues_test, pulsars_target_train, pulsars_target_test = train_test_split(
    pusars.drop(columns="target_class"),
    pusars.target_class,
    test_size=0.25,
    stratify=pusars.target_class
)
```

```python
pulsars_attribues_train.shape
```

```python
svm = SVC(kernel='rbf', gamma=2)
```

```python
svm.fit(pulsars_attribues_train,pulsars_target_train)
```

```python
svm.score(pulsars_attribues_train, pulsars_target_train)
```

```python
svm.score(pulsars_attribues_test, pulsars_target_test)
```

```python
svm = LinearSVC()
```

```python
svm.fit(pulsars_attribues_train,pulsars_target_train)
```

```python
svm.score(pulsars_attribues_train, pulsars_target_train)
```

```python
svm.score(pulsars_attribues_test, pulsars_target_test)
```

```python
svm.coef_
```
